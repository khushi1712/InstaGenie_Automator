## MODULES USED

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
import time
import numpy as np
```

## VISITING THE PAGE

```python
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
driver.get('https://www.instagram.com/')
```

## 1. Login to your Instagram Handle

```python
# username
text_box = driver.find_element(By.NAME,'username')
username = 'SAMPLE'
text_box.send_keys(username)

# password
text_box2 = driver.find_element(By.NAME,'password')
password = 'SAMPLE'
text_box2.send_keys(password)

# for clicking login button
driver.implicitly_wait(3)
driver.find_element(By.XPATH,'//button[contains(@class,"L3NKy")]').click()
print("Login Successful!")
```

## 2. Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”

```python
search = driver.find_element(By.XPATH,'//input[contains(@placeholder,"Search")]')
search.clear()
search_query = 'food'
search.send_keys(search_query)
time.sleep(1)
#sleep is used instead of implicitly wait as,
#here we want program to take pause to load the instagram handles in the search bar...
data = bs(driver.page_source)
data = data.find('div',class_='_abnx _aeul')  #needs to be changes as dynamic class
data = data.find_all('div',role='none')
z=0
for i in data:
    try:
        print(i.a.div.div.find_next_sibling('div').div.div.div.div.string)
        z+=1
    except:
        continue
#error will be generated in case of hastags and locations, 
#so try and except blocks are used to ignore hastags and locations as we only need instagram handles.
```

## 3. Searching and Opening a profile using Insta Handles

```python
# We have to search and open so Delhi
search = driver.find_element(By.XPATH,'//input[contains(@placeholder,"Search")]')
search.clear()
search_query = 'So Delhi'
search.send_keys(search_query)
time.sleep(1)
data = bs(driver.page_source)
data = data.find('div',class_='_abnx _aeul').div # dynamic class keeps changing
link = 'https://www.instagram.com' + data.a['href']
driver.get(link)
```

## 4. Follow/Unfollow given handle

```python
# We need to follow and unfollow 'so delhi'
# Creating a menu driven part to ask to follow/unfollow, and print appropriate messages
data = bs(driver.page_source)
print('-----------------------------')
handle = data.h2.string
print('Instagram Handle:', handle)
buttons = driver.find_elements(By.XPATH,'//button[contains(@class,"_acan")]')
f_button = buttons[1]
try:
    data.find('button',class_='_acan _acap _acas').div.div.string
    f_status = False
except:
    f_status = True
print('Following Status:', f_status)
print('-----------------------------')
print()
print('Enter an option from below: ')
print('1. Follow')
print('2. Unfollow')
print('3. Exit')
while True:
    inp = input('Enter Your Choice: ')
    if inp == '1':
        if f_status is True:
            print('Already Following:',handle)
        else:
            f_button.click()
            f_status = True
            print('Now Following:',handle)
    elif inp == '2':
        if f_status is False:
            print('Your are alerady not following:',handle)
        else:
            f_button.click()
            btn = driver.find_element(By.XPATH,'//button[contains(@class,"_a9-- _a9-_")]')
            btn.click()
            f_status = False
            print('Unfollowed:',handle)
    elif inp == '3':
        break
    else:
        print('ERROR!! \nPlease choose a valid option')
```

## 5. Like/Unlike posts

```python
# opening‘dilsefoodie'
search = driver.find_element(By.XPATH,'//input[contains(@placeholder,"Search")]')
search.clear()
search_query = 'dilsefoodie'
search.send_keys(search_query)
time.sleep(1)
data = bs(driver.page_source)
time.sleep(3)
data = data.find('div',class_='_abnx _aeul').div # dynamic class keeps changing
link = 'https://www.instagram.com' + data.a['href']
driver.get(link)
#link of first 30 posts
time.sleep(5)
driver.execute_script('window.scrollBy(0,10000);')
time.sleep(5)
data = bs(driver.page_source)
lst = data.find_all('div',class_='_ac7v _aang')
links=[]
for i in lst:
    if len(links)==30:
        break
    z = i.div
    for i in range(3):
        links.append('https://www.instagram.com'+z.a['href'])
        z = z.find_next_sibling('div')
print('Post Selected:',len(links))
#liking/Unliking first 30 posts
print('Please choose an option: ')
print('1. Like first 30 posts')
print('2. Dislike first 30 posts')
print('3. To exit')
while True:
    n = input('Your choice: ') 
    if n=='1': #liking
        for i in links:
            driver.get(i)
            time.sleep(1)
            button = driver.find_elements(By.XPATH,'//button[contains(@class,"_abl-")]')
            o = button[1].find_element(By.CSS_SELECTOR,'svg')
            o = o.get_attribute('aria-label')
            if o!='Like':
                print("Already liked")
            else:
                button[1].click()
                print("liked")
    elif n=='2': #Unliking
        for i in links:
            driver.get(i)
            time.sleep(1)
            button = driver.find_elements(By.XPATH,'//button[contains(@class,"_abl-")]')
            o = button[1].find_element(By.CSS_SELECTOR,'svg')
            o = o.get_attribute('aria-label')            
            if o!='Like':
                button[1].click()
                print("Unliked")
            else:
                print("Already Unliked")
    elif n=='3':
        break
    else:
        print('ERROR!! \nPlease choose a valid option')
```
